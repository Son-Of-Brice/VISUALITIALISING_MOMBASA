```python
# Import necessary libraries
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the GeoJSON data
file_path = '/content/drive/MyDrive/DATA/Mombasa.geojson'
mombasa = gpd.read_file(file_path)

# Plot the data
mombasa.plot(figsize=(10, 6), color='lightblue', edgecolor='black')
plt.title("Mombasa Map")
plt.show()
```
